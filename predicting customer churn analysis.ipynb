```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
a=pd.read_csv('Customer Churn.csv')
```

```python
a.drop(['customerID'], axis=1, inplace=True)
a.pop('TotalCharges')
a['OnlineBackup'].unique()
```

```python
#LabelEncoder()
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
a['gender'] = label_encoder.fit_transform(a['gender'])
a['Partner'] = label_encoder.fit_transform(a['Partner'])
a['Dependents'] = label_encoder.fit_transform(a['Dependents'])
a['PhoneService'] = label_encoder.fit_transform(a['PhoneService'])
a['MultipleLines'] = label_encoder.fit_transform(a['MultipleLines'])
a['InternetService'] = label_encoder.fit_transform(a['InternetService'])
a['OnlineSecurity'] = label_encoder.fit_transform(a['OnlineSecurity'])
a['OnlineBackup'] = label_encoder.fit_transform(a['OnlineBackup'])
a['DeviceProtection'] = label_encoder.fit_transform(a['DeviceProtection'])
a['TechSupport'] = label_encoder.fit_transform(a['TechSupport'])
a['StreamingTV'] = label_encoder.fit_transform(a['StreamingTV'])
a['StreamingMovies'] = label_encoder.fit_transform(a['StreamingMovies'])
a['Contract'] = label_encoder.fit_transform(a['Contract'])
a['PaperlessBilling'] = label_encoder.fit_transform(a['PaperlessBilling'])
a['PaymentMethod'] = label_encoder.fit_transform(a['PaymentMethod'])
a['Churn'] = label_encoder.fit_transform(a['Churn'])
```

```python
#separating features(X) and label(y)
a["Churn"] = a["Churn"].astype(int)
Y = a["Churn"].values
X = a.drop(labels = ["Churn"],axis = 1)
```

```python
#train_test_split method
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Import RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
  
# Instantiate the classifier
clf = RandomForestClassifier()
  
# Fit to the training data
clf.fit(X_train, Y_train)
```

```python
# Predict the labels for the test set
y_pred = clf.predict(X_test)
```

```python
# Compute accuracy
from sklearn.metrics import accuracy_score
  
accuracy_score(Y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test, y_pred))
```
